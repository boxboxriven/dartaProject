```python
#delete all observation that have a nan value
cleaned_train = train_df.dropna()
cleaned_train = train_df.reset_index(drop=True)
print("number of observation that have a NAN VALUE for each colmun in cleaned data")
cleaned_train.isnull().sum().sum()
```

```python
#encode all data
enc = OrdinalEncoder()
cleaned_train = enc.fit_transform(train_df)
cleaned_train =  pd.DataFrame(cleaned_train, columns=train_df.columns)
```

```python
# Compute the correlation matrix
correlation_matrix = cleaned_train.corr()
# Set up the matplotlib figure
plt.figure(figsize=(20, 6))
# Draw the heatmap with seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
# Show the plot
plt.show()
```
